```python
# import 해오기
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium
import json
import platform
plt.rc("font", family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 가져올 파일명 리스트로 만들기
gangwon = ["강릉방문자성별연령.csv",
"고성방문자성별연령.csv",
"동해방문자성별연령.csv",
"삼척방문자성별연령.csv",
"속초방문자성별연령.csv",
"양구군방문자성별연령.csv",
"양양방문자성별연령.csv",
"영월방문자성별연령.csv",
"원주방문자성별연령.csv",
"인제방문자성별연령.csv",
"정선군방문자성별연령.csv",
"철원방문자성별연령.csv",
"춘천방문자성별연령.csv",
"태백방문자성별연령.csv",
"평창군방문자성별연령.csv",
"홍천군방문자성별연령.csv",
"화천군방문자성별연령.csv",
"횡성군방문자성별연령.csv"]
```

```python
# 파일명을 데이터 프레임화해서 지역이름만 뽑아오기
df = pd.DataFrame()
df["이름"] = gangwon
localname = df["이름"].str[0:2]
# 지역이름을 뽑아서 리스트로 만든걸 또새로운 리스트를 만들어 나중에 for문 돌릴때 name[0] = 새로운 데이터 프레임으로 만들수 있게 하기
name = localname.copy()
```

```python
for i in range(18):
    # 파일을 경로를 불러오기
    file_path="./강원도 지역별 관광객 연령/{}".format(gangwon[i])
    # name[0] 을 새로운 데이터프레임으로 만들기
    name[i] = pd.DataFrame()
    # 엑셀 파일을 데이터 프레임에 넣기
    name[i] = pd.read_csv(file_path, encoding="euc-kr")
    # 남자와 여자를 합해 총 방문자 수를 만들기
    name[i]["방문자수"] = name[i]["방문자 수(남성)"] + name[i]["방문자 수(여성)"]
    # 지역별 컬럼 만들기
    name[i].columns =  ['{} 방문자 연령'.format(localname[i]),
                        '{} 방문자 수(남성)'.format(localname[i]), 
                        '{} 방문자 비율(남성)'.format(localname[i]),
                        '{} 방문자 수(여성)'.format(localname[i]),
                        '{} 방문자 비율(여성)'.format(localname[i]),
                        '{} 방문자수'.format(localname[i])] 
```

```python
name[0].columns
```

```python
# 시각화를 y축에 들어갈 리스트 만들기
ti =[]
for i in range(18):
    ti.append('{} 방문자수'.format(localname[i]))
```

```python
# 지역별로 구분해 놓은 컬럼을 다 합치기
potato_sum = pd.DataFrame()
for i in range(0,18):
    potato_sum = pd.concat([potato_sum ,name[i]], axis=1)
# 합친걸 엑셀로 저장
potato_sum.to_excel("./강원도 지역별 관광객 연령/강원도 전지역통합.xlsx")
```

```python
potato_sum = pd.read_excel("./강원도 지역별 관광객 연령/강원도 전지역통합.xlsx")
```

```python
# 지역별 연령별 방문자 수 시각화
for i in range(18):
    plt.figure(figsize=(13,8))
    plt.plot(potato_sum["고성 방문자 연령"],potato_sum['{} 방문자수'.format(localname[i])])
    plt.title("{}".format(localname[i]))
    plt.savefig('연령별 방문자 수 시각화{}.png'.format(localname[i]))
    plt.show()
```

```python
# 지역별로 나눈것을 하나의 맵에 그리기
plt.figure(figsize=(13,8))
for i in range(18):
    plt.plot(potato_sum["고성 방문자 연령"],potato_sum['{} 방문자수'.format(localname[i])], label=localname[i])
plt.title("지역별 연령별 방문자수")
plt.legend()
plt.savefig('지역별 연령별 방문자수.png')
plt.show()
```

```python
# 지역별 연령별 성별 방문자 수 시각화
for i in range(18):
    plt.figure(figsize=(13,8))
    plt.plot(potato_sum["고성 방문자 연령"],potato_sum['{} 방문자 수(여성)'.format(localname[i])],label="여성" )
    plt.plot(potato_sum["고성 방문자 연령"],potato_sum['{} 방문자 수(남성)'.format(localname[i])], label="남성")
    plt.title("{}".format(localname[i]))
    plt.legend()
    plt.savefig('연령별 성별 방문자 수 시각화{}.png'.format(localname[i]))
    plt.show()
```

```python
# 지역별 연령별 방문자 수 시각화 
potato_sum.plot.bar(x= "고성 방문자 연령", y= ti, figsize = (20,10), width=0.9)
plt.grid()
```
