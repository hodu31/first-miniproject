```python
# 음식 트렌드 데이터에 대한 컬럼정의서 로드
file_path = "./gandata/food/지역별 음식점 소비기반 트렌드_컬럼정의서.xls"
df_master = pd.read_excel(file_path)
df_master

# 컬럼 영문명과 한글명을 딕셔너리로 매핑
a = dict(zip(df_master["컬럼영문명"],df_master["컬럼한글명"]))

# 컬럼명 변경
df = df.rename(columns=a)
```

```python
df
```

```python
df["시도명"].unique()
```

```python
df_new = df.iloc[:,[0,1,4,6]].copy()
```

```python
condition = (df_new["시도명"] == "대구광역시")
df_dae= df_new[condition == True]
```

```python
dae_pivot = df_dae.pivot_table(values="음식시설수",
                       index="시군구명",
                       columns="음식시설명",
                       aggfunc="mean")
```

```python
sns.heatmap(dae_pivot, cmap="rocket_r", annot=True)
```

```python
condition = (df_new["시도명"] == "강원도")
df_gang= df_new[condition == True]
```

```python
gang_pivot = df_gang.pivot_table(values="음식시설수",
                       index="시군구명",
                       columns="음식시설명",
                       aggfunc="mean")
```

```python
sns.heatmap(gang_pivot, cmap="rocket_r", annot=True)
```
