```python
# import 해오기
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium
import json
import platform
plt.rc("font", family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False
```

# A. 데이터 통합

```python
# 위치한 폴더 경로
folder_path = "./data/4년치 강원도 소셜미디어 관련 데이터"

# 폴더 내의 파일 목록을 가져옴
file_list = os.listdir(folder_path)
```

```python
# 폴더내 파일명 불러오기
file_list
```

```python
# 여행유형 트렌트 키워드 파일명 불러오기
filtered_list = [file for file in file_list if '트랜드키워드' in file]
```

```python
filtered_list, len(filtered_list)
```

```python
area_number =[]
for i in range(18):
    area_number.append("{}".format(i) + "번:" +filtered_list[i][4:7])
```

```python
area_number
```

```python
df = filtered_list.copy()
```

```python
for i in range(len(filtered_list)):
    # 파일을 경로를 불러오기
    file_path="./data/4년치 강원도 소셜미디어 관련 데이터/{}".format(filtered_list[i])
    
    # name[0] 을 새로운 데이터프레임으로 만들기
    df[i] = pd.DataFrame()
    
    # 엑셀 파일을 데이터 프레임에 넣기
    df[i] = pd.read_csv(file_path, encoding="euc-kr")
    
    # 년도별 지역별 컬럼 만들기
    df[i].columns =  ['{}년도{} 트랜드키워드 순위'.format(filtered_list [i][0:4], filtered_list[i][4:7]),
                      '{}년도{} 레저 분류'.format(filtered_list [i][0:4], filtered_list[i][4:7]), 
                      '{}년도{} 관련키워드'.format(filtered_list [i][0:4], filtered_list[i][4:7]),
                      '{}년도{} 언급건수'.format(filtered_list [i][0:4], filtered_list[i][4:7])]
```

```python
df
```

```python
# 데이터 통합
potato_sum = pd.DataFrame()
for i in range(len(filtered_list)):
    potato_sum = pd.concat([potato_sum ,df[i]], axis = 1)
potato_sum
```

```python
# 통합한 데이터 저장하기
potato_sum.to_csv("./data/4년치 강원도 소셜미디어 관련 데이터/2019~2022 여행유형 트렌드 키워드.csv", index = False)
```

```python
# 10위까지만 사용
kangwon = potato_sum[:10]
kangwon
```

# B. 데이터 시각화

```python
# 시군구명만 담긴 리스트 만들기
a = filtered_list[:18]
a_name = []
for i in range(18):
    a_name.append(a[i][4:7])
a_name
```

```python
강릉 = pd.DataFrame()
for i in range(0, 55, 18):
    강릉 = pd.concat([강릉, df[i]], axis=1)
강릉
```

```python
강릉 = 강릉.iloc[:30]
강릉
```

```python
강릉_pi= 강릉.pivot_table(values="2019년도강릉시 언급건수",
              index="2019년도강릉시 관련키워드")

plt.plot(강릉_pi)
```

### 완성!

```python
for i in range(18):
    plt.figure()
    kangwon_2019 = kangwon.pivot_table(values = "2019년도{} 언급건수".format(a_name[i]),
                                       index = "2019년도{} 관련키워드".format(a_name[i])).sort_values("2019년도{} 언급건수".format(a_name[i]), ascending = False).reset_index()
    
    kangwon_2020 = kangwon.pivot_table(values = "2020년도{} 언급건수".format(a_name[i]),
                                       index = "2020년도{} 관련키워드".format(a_name[i])).sort_values("2020년도{} 언급건수".format(a_name[i]), ascending = False).reset_index()
    
    kangwon_2021 = kangwon.pivot_table(values = "2021년도{} 언급건수".format(a_name[i]),
                                       index = "2021년도{} 관련키워드".format(a_name[i])).sort_values("2021년도{} 언급건수".format(a_name[i]), ascending = False).reset_index()
    
    kangwon_2022 = kangwon.pivot_table(values = "2022년도{} 언급건수".format(a_name[i]),
                                       index = "2022년도{} 관련키워드".format(a_name[i])).sort_values("2022년도{} 언급건수".format(a_name[i]), ascending = False).reset_index()
    
    # 그래프
    plt.subplot(2, 2, 1)
    plt.rcParams["figure.figsize"] = (10, 10)
    plt.pie(kangwon_2019["2019년도{} 언급건수".format(a_name[i])], autopct = "%.1f%%", startangle = 90, counterclock = False, pctdistance = 0.7)
    plt.legend(kangwon_2019["2019년도{} 관련키워드".format(a_name[i])], bbox_to_anchor = (1, 1), prop={'size': 7})
    plt.title("2019년")
    
    plt.subplot(2, 2, 2)
    plt.rcParams["figure.figsize"] = (10, 10)
    plt.pie(kangwon_2020["2020년도{} 언급건수".format(a_name[i])], autopct = "%.1f%%", startangle = 90, counterclock = False, pctdistance = 0.7)
    plt.legend(kangwon_2020["2020년도{} 관련키워드".format(a_name[i])], bbox_to_anchor = (1, 1), prop={'size': 7})
    plt.title("2020년")
    
    plt.subplot(2, 2, 3)
    plt.rcParams["figure.figsize"] = (10, 10)
    plt.pie(kangwon_2021["2021년도{} 언급건수".format(a_name[i])], autopct = "%.1f%%", startangle = 90, counterclock = False, pctdistance = 0.7)
    plt.legend(kangwon_2021["2021년도{} 관련키워드".format(a_name[i])], bbox_to_anchor = (1, 1), prop={'size': 7})
    plt.title("2021년")
    
    plt.subplot(2, 2, 4)
    plt.rcParams["figure.figsize"] = (10, 10)
    plt.pie(kangwon_2022["2022년도{} 언급건수".format(a_name[i])], autopct = "%.1f%%", startangle = 90, counterclock = False, pctdistance = 0.7)
    plt.legend(kangwon_2022["2022년도{} 관련키워드".format(a_name[i])], bbox_to_anchor = (1, 1), prop={'size': 7})
    plt.title("2022년")
    
    plt.suptitle("연도별 {} 여행 트렌드 키워드".format(a_name[i]))
    plt.tight_layout()
```
