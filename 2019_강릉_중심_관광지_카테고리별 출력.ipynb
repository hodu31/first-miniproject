```python
folder_path = "./2019_2022_강원도_카테고리별_관광인프라/중심관광지"

file_list = os.listdir(folder_path)
df=[]

df_2019=pd.DataFrame()
### 18->16->15
for i in range(0,18,1):
    # 파일명에서 시군명 추출
    sigoon=file_list[i][4:7]
    # 각 파일경로 지정
    file_path=("./2019_2022_강원도_카테고리별_관광인프라/중심관광지/2022{}중심 관광지.csv".format(sigoon))
    # df에 각각의 데이터프레임명 추가하기
    df.append(pd.read_csv(file_path, encoding="euc-kr"))
    del(df[i]["중심카테고리 명_대"])
```

```python
lst_area = []
lst_area = df[0]["분류"].unique()
lst_area
```

```python
lst_file = []
for i in range(18):
     lst_file.append(df[i]["분류"].unique())
lst_file
```

```python
# 관광 종류 별로 좌표갑 데이터 프레임에 담기
b =[]
for i in range(18):
    a= []
    lst_file.append(df[i]["분류"].unique())
    for p in range(len(lst_file[i])):
        condition = df[i]["분류"] == lst_file[i][p]
        a.append(df[i][condition == True].reset_index(drop=True))
    b.append(a)
```

```python
qwe = []
for i in range(18):
    qwe.append(b[i][0]["주소"][0].split()[1])
qwe
```

```python
import folium

color_list = ['#0a676c', '#70a3bd', '#c16e2e', "#89cff0", "#790b9e", "#e51111", "#6c78d7", "#403157"]

gw_list = []
gw_map = folium.Map(location=[37.882531, 127.7291452], tiles="CartoDB positron", zoom_start=7)

# Create a FeatureGroup object
layer_control = folium.LayerControl(collapsed=True)

for i in range(18):
    for p in range(len(lst_file[i])):
        layer = folium.FeatureGroup(name=f"Layer {i+1}" + ":" + "{}".format(qwe[i]) + "{}".format(lst_file[i][p]), show=False)
        for x in range(len(b[i][p])):
            lat = b[i][p].loc[x, "중심 POI Y 좌표"]
            lng = b[i][p].loc[x, "중심 POI X 좌표"]
            rank = b[i][p].loc[x, "순위"]
            name = b[i][p].loc[x, "관광지명"]
            if rank <= 20:
                marker_weight = 10
            elif rank <= 40:
                marker_weight = 8
            elif rank <= 60:
                marker_weight = 6
            elif rank <= 80:
                marker_weight = 4
            elif rank <= 100:
                marker_weight = 2

            colorcolor = color_list[p]
            folium.CircleMarker(location=[lat, lng],
                                fill_color=colorcolor,
                                fill_opacity=1,
                                color=colorcolor,
                                weight=marker_weight,
                                radius=marker_weight,
                                popup=name).add_to(layer)

        layer.add_to(gw_map)

layer_control.add_to(gw_map)

gw_map.save("map.html")
```
