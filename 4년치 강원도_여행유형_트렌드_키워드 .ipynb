```python
# 위치한 폴더 경로
folder_path = "./4년치 강원도 소셜미디어 관련 데이터"

# 폴더 내의 파일 목록을 가져옴
file_list = os.listdir(folder_path)
```

```python
# 여행유형 트렌트 키워드 파일명 불러오기
filtered_list = [file for file in file_list if '트랜드키워드' in file]
```

```python
area_number =[]
for i in range(18):
    area_number.append("{}".format(i) + "번:" +filtered_list[i][4:7])
```

```python
df = filtered_list.copy()
```

```python
for i in range(len(filtered_list)):
    # 파일을 경로를 불러오기
    file_path="./4년치 강원도 소셜미디어 관련 데이터/{}".format(filtered_list[i])
    
    # name[0] 을 새로운 데이터프레임으로 만들기
    df[i] = pd.DataFrame()
    
    # 엑셀 파일을 데이터 프레임에 넣기
    df[i] = pd.read_csv(file_path, encoding="euc-kr")
    
    # 년도별 지역별 컬럼 만들기
    df[i].columns =  ['{}년도{} 트랜드키워드 순위'.format(filtered_list [i][0:4], filtered_list[i][4:7]),
                      '{}년도{} 레저 분류'.format(filtered_list [i][0:4], filtered_list[i][4:7]), 
                      '{}년도{} 관련키워드'.format(filtered_list [i][0:4], filtered_list[i][4:7]),
                      '{}년도{} 언급건수'.format(filtered_list [i][0:4], filtered_list[i][4:7])]
```
