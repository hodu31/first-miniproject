```python
#import 해오기
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium
import json
import platform
plt.rc("font", family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 여가문화 데이터 로드
df = pd.DataFrame()
for i in range(1,79):
    file_path = "./data/leisure/여가문화 ({}).csv".format(i)
    temp = pd.read_csv(file_path)
    df = pd.concat([df , temp])
```

```python
# 여가문화 시간 데이터에 대한 컬럼정의서 로드
file_path = "./data/leisure/하루 평균 여가문화 시간 및 사용 비중_컬럼정의서.xls"
df_master = pd.read_excel(file_path)
df_master

# 컬럼 영문명과 한글명을 딕셔너리로 매핑
a = dict(zip(df_master["컬럼영문명"],df_master["컬럼한글명"]))

# 컬럼명 변경
df = df.rename(columns=a)
```

```python
# 성별 컬럼 추가
df["성별"] = df["성별구분코드"].map({"M":"남자", "F":"여자"})
df["조사시작일자"] = df["조사시작일자"].astype(str)

yy=[]
mm=[]
dd=[]
# 조사시작일자를 연, 월, 일로 분리하여 컬럼 추가
for val in df["조사시작일자"]:
    yy.append(val[:4])
    mm.append(val[4:6])
    dd.append(val[6:])
df["년"] = yy
df["월"] = mm
df["일"] = dd


df["년"] = df["년"].astype(int)
df["월"] = df["월"].astype(int)
df["일"] = df["일"].astype(int)
```

```python
# 성별구분코드 컬럼 삭제
df = df.drop("성별구분코드", axis=1)
```

```python
# 비율 대신 시간으로 바꾸는새로운 DataFrame 생성
df["레저시간휴식오락시간값"] = df["레저시간휴식오락사용비율"] / 100 * df["1주총레저시간값"]
df["레저시간취미사용시간값"] = df["레저시간취미사용비율"] / 100 * df["1주총레저시간값"]
df["레저시간본인계발시간값"] = df["레저시간본인계발사용비율"] / 100 * df["1주총레저시간값"]
df["레저시간대인관계교제시간값"] = df["레저시간대인관계교제사용비율"] / 100 * df["1주총레저시간값"]
df["레저시간기타시간값"] = df["레저시간기타사용비율"] / 100 * df["1주총레저시간값"]
```
